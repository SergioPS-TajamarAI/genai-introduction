```python
import pandas as pd
# Importing the dataset
# The dataset is a CSV file containing information about houses in Spain
df = pd.read_csv(dataset_path)
print(df.columns.tolist())
print(df.columns)
```

```python
# Select relevant columns
df = df[['price','m2_real', 'bath_num', 'room_num', 'house_type', 
         'garage', 'garden', 'swimming_pool', 'terrace', 
        'air_conditioner', 'heating', 'loc_city']]
df.sample(20)
```

```python
df = df[~df['house_type'].str.contains('alquiler', case=False, na=False)]
df.reset_index(drop=True, inplace=True)
```

```python
df = df[df['house_type'] != 'house_type']
df.reset_index(drop=True, inplace=True)
```

```python
house_types = df['house_type'].unique().tolist()
print(house_types)
```

```python
# Obtencion de dummies para house_type
house_type_dummies = pd.get_dummies(df['house_type'], prefix='house_type').astype(int)
df = pd.concat([df, house_type_dummies], axis=1)

# Eliminar la columna house_type
df.drop(columns=['house_type'], inplace=True)
```

```python
# Convertir las columnas de tipo object a tipo int
df['garage'] = df['garage'].apply(lambda x: 0 if str(x).lower() == 'nan' else 1)
```

```python
# Eliminacion de columnas innecesarias
df['heating'] = df['heating'].apply(lambda x: 0 if str(x).lower() == 'nan' else 1)
```

```python
import matplotlib.pyplot as plt

# Convert columns to numeric
df[['price', 'm2_real', 'bath_num', 'room_num']] = df[['price', 'm2_real', 'bath_num', 'room_num']].apply(pd.to_numeric, errors='coerce')

# Mostrar un boxplot para identificar outliers
plt.figure(figsize=(12, 6))
df.boxplot(column=['price', 'm2_real', 'bath_num', 'room_num'])
plt.title("Boxplot para identificar outliers")
plt.show()

# Eliminar outliers utilizando el rango intercuartílico (IQR)
def remove_outliers(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

columns_to_check = ['price', 'm2_real', 'bath_num', 'room_num']
df = remove_outliers(df, columns_to_check)

# Resetear el índice después de eliminar los outliers
df.reset_index(drop=True, inplace=True)
```

```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras import regularizers

# Cargar el dataset (suponiendo que 'df' ya está cargado en el entorno)
df['price'] = df['price'].astype(float)
df['m2_real'] = df['m2_real'].astype(float)
df['bath_num'] = df['bath_num'].astype(float)
df['room_num'] = df['room_num'].astype(float)

# Selección de las características (X) y la variable objetivo (y)
X = df.drop(columns=['price']).apply(pd.to_numeric, errors='coerce').fillna(0).values
y = df['price'].values
```

```python
# Función para crear el modelo sin optimización
def create_initial_model():
    model = Sequential([
        Dense(12, input_dim=X.shape[1], activation='relu'),
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),
        Dense(1, activation='linear')
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

# Crear y entrenar el modelo inicial
initial_model = create_initial_model()
initial_model.fit(X, y, epochs=10, batch_size=32, verbose=1)

# Realizar una predicción inicial para verificar el modelo
sample_pred = initial_model.predict(X[:5])
print("Predicciones iniciales:", sample_pred.flatten())
```

```python
# Definir el número de splits para K-Fold
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Función para crear el modelo con optimización
def create_optimized_model(optimizer, learning_rate=0.001, dropout_rate=0.2, l2_reg=0.01):
    model = Sequential([
        Dense(12, input_dim=X.shape[1], activation='relu', kernel_regularizer=regularizers.l2(l2_reg)),
        Dropout(dropout_rate),
        Dense(64, activation='relu', kernel_regularizer=regularizers.l2(l2_reg)),
        Dropout(dropout_rate),
        Dense(32, activation='relu', kernel_regularizer=regularizers.l2(l2_reg)),
        Dense(1, activation='linear')
    ])
    
    model.compile(optimizer=optimizer(learning_rate=learning_rate), loss='mse', metrics=['mae'])
    return model

# Evaluación del modelo optimizado con validación cruzada
def evaluate_model(optimizer, learning_rate=0.001, dropout_rate=0, l2_reg=0.01):
    results = []
    for train_index, val_index in kf.split(X):
        X_train, X_val = X[train_index], X[val_index]
        y_train, y_val = y[train_index], y[val_index]

        # Crear el modelo optimizado
        model = create_optimized_model(optimizer, learning_rate, dropout_rate, l2_reg)

        # Entrenar el modelo optimizado
        model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=32, verbose=0)

        # Predicciones
        y_pred = model.predict(X_val).flatten()

        # Calcular métricas
        mae = mean_absolute_error(y_val, y_pred)
        rmse = np.sqrt(mean_squared_error(y_val, y_pred))
        r2 = r2_score(y_val, y_pred)

        results.append({'MAE': mae, 'RMSE': rmse, 'R2': r2})
    
    return results
```

```python
# Definir los optimizadores y sus parámetros
optimizers = {
    'Adam': Adam,
    'SGD': SGD
}

learning_rates = [0.001, 0.0001]
dropout_rates = [0.0, 0.2]
l2_regs = [0.0, 0.01]

# Almacenar resultados para cada combinación de hiperparámetros
experiment_results = {}

for optimizer_name, optimizer in optimizers.items():
    for lr in learning_rates:
        for dropout in dropout_rates:
            for l2_reg in l2_regs:
                print(f"Evaluando: {optimizer_name} | LR: {lr} | Dropout: {dropout} | L2 Reg: {l2_reg}")
                
                # Evaluar el modelo con estos hiperparámetros
                results = evaluate_model(optimizer, learning_rate=lr, dropout_rate=dropout, l2_reg=l2_reg)
                experiment_results[(optimizer_name, lr, dropout, l2_reg)] = results
```

```python
# Comparar resultados obtenidos
max_y = np.max(y)  # Valor máximo de referencia en y

for params, results in experiment_results.items():
    optimizer_name, lr, dropout, l2_reg = params
    avg_mae = np.mean([res['MAE'] for res in results])
    avg_rmse = np.mean([res['RMSE'] for res in results])
    avg_r2 = np.mean([res['R2'] for res in results])
    
    # Calcular porcentajes
    mae_percentage = (avg_mae / max_y) * 100
    rmse_percentage = (avg_rmse / max_y) * 100

    print(f"\nResultados para {optimizer_name} | LR: {lr} | Dropout: {dropout} | L2: {l2_reg}")
    print(f"Average MAE: {avg_mae:.4f} ({mae_percentage:.2f}%)")
    print(f"Average RMSE: {avg_rmse:.4f} ({rmse_percentage:.2f}%)")
    print(f"Average R2: {avg_r2:.4f}")
```

```python
# Selección del mejor modelo basado en resultados previos
best_optimizer = Adam  # Suponiendo que Adam es el mejor optimizador
best_lr = 0.001  # Ejemplo de mejor tasa de aprendizaje
best_dropout = 0.0  # Mejor tasa de dropout
best_l2_reg = 0.01  # Mejor regularización L2

# Crear y entrenar el modelo final con los mejores hiperparámetros
final_model = create_optimized_model(best_optimizer, learning_rate=best_lr, dropout_rate=best_dropout, l2_reg=best_l2_reg)
final_model.fit(X, y, epochs=50, batch_size=32, verbose=1)

# Evaluación final en el conjunto de test
final_predictions = final_model.predict(X).flatten()
final_mae = mean_absolute_error(y, final_predictions)
final_rmse = np.sqrt(mean_squared_error(y, final_predictions))
final_r2 = r2_score(y, final_predictions)

# Calcular porcentajes
max_y = np.max(y)
mae_percentage = (final_mae / max_y) * 100
rmse_percentage = (final_rmse / max_y) * 100

print("\nEvaluación en conjunto de test:")
print(f"Final MAE: {final_mae:.4f} ({mae_percentage:.2f}%)")
print(f"Final RMSE: {final_rmse:.4f} ({rmse_percentage:.2f}%)")
print(f"Final R2: {final_r2:.4f}")
```
